```python
import pandas as pd
import random

students = []

for i in range(1, 981):

    student_id = f"STU{i:04d}"
    name = f"Student_{i}"
    email = f"student{i}@gmail.com"

    classes_attended = random.randint(0, 40)

    attendance_percentage = round(
        (classes_attended / 40) * 100,
        2
    )

    certificate_status = (
        "Eligible"
        if classes_attended >= 32
        else "Not Eligible"
    )

    students.append([
        student_id,
        name,
        email,
        classes_attended,
        attendance_percentage,
        certificate_status
    ])

final_df = pd.DataFrame(
    students,
    columns=[
        "Student_ID",
        "Student_Name",
        "Email",
        "Classes_Attended",
        "Attendance_Percentage",
        "Certificate_Status"
    ]
)

final_df.to_excel(
    "Certificate_Eligibility_Report.xlsx",
    index=False
)

print("980 Student Report Generated Successfully")

from IPython.display import display

pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

display(final_df)
```

```python
import matplotlib.pyplot as plt

# Count Eligible and Not Eligible students
status_counts = final_df["Certificate_Status"].value_counts()

# Create graph
plt.figure(figsize=(6, 4))
status_counts.plot(kind="bar")

plt.title("Certificate Eligibility Report")
plt.xlabel("Certificate Status")
plt.ylabel("Number of Students")

plt.tight_layout()

# Save graph as image
plt.savefig("certificate_eligibility_graph.png")

plt.show()

print("Graph saved as: certificate_eligibility_graph.png")
```

```python
import matplotlib.pyplot as plt

status_counts = final_df["Certificate_Status"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    status_counts,
    labels=status_counts.index,
    autopct="%1.1f%%"
)

plt.title("Certificate Eligibility Distribution")

plt.savefig("certificate_eligibility_pie_chart.png")

plt.show()

print("Pie chart saved as: certificate_eligibility_pie_chart.png")
```
